```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('financial_sample.csv')
df
```

```python
df.columns = df.columns.str.strip()
```

```python
df['Units Sold'] = pd.to_numeric(df['Units Sold'].replace({'\$': '', ',': ''}, regex=True), errors='coerce')
df['Sales'] = pd.to_numeric(df['Sales'].replace({'\$': '', ',': ''}, regex=True), errors='coerce')
df['Profit'] = pd.to_numeric(df['Profit'].replace({'\$': '', ',': ''}, regex=True), errors='coerce')
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
```

```python
summary_stats = df[['Units Sold', 'Sales', 'Profit']].describe()
```

```python
summary_stats
```

```python
sales_by_segment_country = df.groupby(['Segment', 'Country'])[['Sales', 'Profit']].sum().sort_values(by='Sales', ascending=False)
```

```python
# Grouping by Segment and Country to observe sales and profit distributions
sales_by_segment_country
```

```python
# Correlation analysis between financial variables
correlation_matrix = df[['Units Sold', 'Sales', 'Profit']].corr()
```

```python
correlation_matrix
```

```python
sales_trends = df.groupby(['Year', 'Month Name'])[['Sales']].sum()
```

```python
 #Time-series analysis to check sales trends over the years
sales_trends
```
